## GIS data on Streetlights in LA

#### API Connection to GIS Data

Utilize the following code to create a connection with the API. Please replace *'User-agent': 'cbarry@deloitte.com'* with *'User-agent': **'your-email-here**'*, and make sure you have all of the libraries used below installed.

```python
import numpy as np
import pandas as pd
import json
import requests
from datetime import datetime
```

```python
root = "https://dpw.gis.lacounty.gov"
endpoint = "/dpw/rest/services/PW_Open_Data/MapServer/5/query?where=1%3D1&outFields=*&outSR=4326&f=json"
```

```python
r = requests.get(root + endpoint, headers={'User-agent': 'cbarry@deloitte.com'})
r
```

```python
gis_df = pd.json_normalize(json.loads(r.text), record_path = ['features'])
```

```python
gis_df.head()
```
